```python
tasks = []

def display_menu():
    """Week 5 – Loop control and menu display"""
    print("\nTask Manager Menu:")
    print("1. Add task")
    print("2. List tasks")
    print("3. Remove task")
    print("4. Quit")

def add_task(tasks_list, text):
    """
    Adds a task to the list.
    Week 4: Function with parameters
    Week 6: Defensive programming (input validation)
    Week 9: Doctest placeholder
    >>> tasks_list = []
    >>> add_task(tasks_list, "Buy milk")
    Added: Buy milk
    >>> tasks_list
    ['Buy milk']
    >>> add_task(tasks_list, "")
    Error: Task cannot be empty
    >>> tasks_list
    ['Buy milk']
    """
    if text.strip() == "":
        print("Error: Task cannot be empty")
        return
    tasks_list.append(text.strip())
    print("Added:", text.strip())

def list_tasks(tasks_list):
    """
    Lists all tasks with their index.
    Week 4: Function with parameter
    Week 5: Loop pattern
    """
    if not tasks_list:
        print("No tasks available.")
    else:
        print("\nYour Tasks:")
        for i, task in enumerate(tasks_list):
            print(f"{i + 1}: {task}")

def remove_task(tasks_list, index):
    """
    Removes a task by index.
    Week 4: Function with parameters
    Week 6: Defensive programming (index validation)
    Week 8: Basic try/except handled in calling code (get_remove_index)
    Week 9: Doctest placeholder
    >>> tasks_list = ["a", "b", "c"]
    >>> remove_task(tasks_list, 1)
    Removed: b
    >>> tasks_list
    ['a', 'c']
    >>> remove_task(tasks_list, 5)
    Invalid index!
    >>> tasks_list
    ['a', 'c']
    >>> remove_task(tasks_list, -1)
    Invalid index!
    >>> tasks_list
    ['a', 'c']
    """
    if 0 <= index < len(tasks_list):
        removed_task = tasks_list.pop(index)
        print("Removed:", removed_task)
    else:
        print("Invalid index!")

def get_choice():
    """
    Gets user menu choice.
    Week 4: Simple function
    Week 6: Basic input handling
    """
    choice = input("Enter your choice: ").strip()
    return choice

def get_task_text():
    """
    Gets task text input from user.
    Week 4: Simple function
    """
    text = input("Enter task description: ").strip()
    return text

def get_remove_index():
    """
    Gets and validates index for removal.
    Week 4: Simple function
    Week 8: Basic try/except
    """
    raw_input = input("Enter task number to remove: ").strip()
    try:
        idx = int(raw_input) - 1  # Convert to 0-based index
        return idx
    except ValueError:
        print("Invalid input. Please enter a number.")
        return None

# --- Main Program Loop ---
# Week 5: While loop pattern

print("Starting Task Manager...")

while True:
    display_menu()
    user_choice = get_choice()

    if user_choice == "1":
        task_description = get_task_text()
        add_task(tasks, task_description)
    elif user_choice == "2":
        list_tasks(tasks)
    elif user_choice == "3":
        index_to_remove = get_remove_index()
        if index_to_remove is not None:
            remove_task(tasks, index_to_remove)
    elif user_choice == "4":
        print("Goodbye!")
        break
    else:
        print("Invalid option. Please try again.")

print("Task Manager finished.")

# Course References in this section:
# Week 3: Followed planning structure from pseudocode
# Week 4: Implemented functions with parameters/returns
# Week 5: Used while loop for menu and for loop for listing
# Week 6: Added basic input validation (empty task, index range)
# Week 8: Used try/except for index input conversion
# Week 9: Included doctest examples within function docstrings
```
